```python
# inspired from: http://alexperrier.github.io/jekyll/update/2015/09/04/topic-modeling-of-twitter-followers.html

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
%load_ext autoreload
%autoreload 2

# Basics
import pickle, codecs, csv, copy, string, re
from collections import OrderedDict
import numpy as np
from gensim import corpora, models
import pyLDAvis.gensim
import pandas as pd
import seaborn as sns
#sns.set(color_codes=True)
sns.set_context("paper")
```

```python
n_topics = 10
```

```python
# A1
# Load the corpus and dictionary

corpus = corpora.MmCorpus('txt_tm_models/A1_ec_hist_%d.mm'%n_topics)
dictionary = corpora.Dictionary.load('txt_tm_models/A1_ec_hist_%d.dict'%n_topics)

# First LDA model with 10 topics, 10 passes, alpha = 0.001
lda = models.LdaModel.load('txt_tm_models/A1_ec_hist_%d.lda'%n_topics)
followers_data =  pyLDAvis.gensim.prepare(lda, corpus, dictionary)
pyLDAvis.display(followers_data)
```
